```
#palindrome
n = int(input("enter a value"))
rev = 0
temp = n
while n != 0 :
  r = n % 10 
  rev = rev * 10 + r
  n = n//10
if temp == rev :
  print("is a palindrome")
else :
  print(" is not a palindrome")
```

```
def even(n):
    s = 0
    for i in range (1, n+1):
        if i % 2 == 0 :
          s = s + i
          i = i + 1  
    return s
```

```
def fact(b):
    f = 1
    for i in range(1,b+1):
        f = f * i
    return f
```
